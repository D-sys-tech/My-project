```python
#percentage of missing values in the aviation dataset.
def missing_values(df_aviation_data_2):
  miss = df_aviation_data_2.isnull().sum()
  percentage_miss = (df_aviation_data_2.isnull().sum() / len (df_aviation_data_2)) * 100
  missing = pd.DataFrame({"missing_values": miss, "Percentage%": percentage_miss})
  missing = missing[missing["missing_values"] > 0]
  return missing

missing_data = missing_values(df_aviation_data)
print(missing_data)
```

## Data cleaning for the aviation data
Data cleaning for df_aviation_data_2
1.Get rid of;latitude,longitude,Airport. code,Airport. Name,Far description,schedule,Air carrier since they have the highest pecentage of missing values.
2.Fill missing Total.Fatal.Injuries,Total.Serious.Injuries,Total.Minor.Injuries, Total.Uninjured with the mean values.



```python
#Drop columns with the highest percentage of missing data.
drop_columns = ["Latitude","Longitude","Airport.Code","Airport.Name","FAR.Description","Schedule","Air.carrier"]
df_aviation_data_2_cleaned = df_aviation_data_2.drop(columns=drop_columns)
df_aviation_data_2_cleaned
```

```python
#Fill in missing injury values
df_aviation_data_2_cleaned.loc[:,["Location","Country","Weather.Condition","Report.Status"]].fillna("Unknown", inplace=True)
df_aviation_data_2_cleaned.loc[:,["Total.Fatal.Injuries","Total.Serious.Injuries","Total.Minor.Injuries","Total.Uninjured"]].fillna(0,inplace=True)
print(df_aviation_data_2_cleaned.isnull().sum())
```

```python
print(df_aviation_data_2_cleaned.describe(include="all"))
```

```python
# Frequency of incidents by Aircraft Make and Model
make_model_incidents = df_aviation_data_2_cleaned.groupby(['Make', 'Model']).size().sort_values(ascending=False).head(10)
print(make_model_incidents)
```

```python
# Incidents involving fatalities or serious injuries
fatal_injuries = df_aviation_data_2_cleaned[df_aviation_data_2_cleaned['Total.Fatal.Injuries'] > 0]
serious_injuries = df_aviation_data_2_cleaned[df_aviation_data_2_cleaned['Total.Serious.Injuries'] > 0]

fatal_injuries_by_make = fatal_injuries.groupby('Make').size().sort_values(ascending=False).head(10)
serious_injuries_by_make = serious_injuries.groupby('Make').size().sort_values(ascending=False).head(10)

print("Fatal Injuries by Aircraft Make:\n", fatal_injuries_by_make)
print("Serious Injuries by Aircraft Make:\n", serious_injuries_by_make)
```

```python
# Analyze the frequency of substantial damage or destroyed aircraft
damage_incidents = df_aviation_data_2_cleaned[df_aviation_data_2_cleaned['Aircraft.damage'] == 'Destroyed']
damage_by_make = damage_incidents.groupby('Make').size().sort_values(ascending=False).head(10)

print("Destroyed Aircraft by Make:\n", damage_by_make)
```

```python
# Analyze incidents by weather conditions
weather_incidents = df_aviation_data_2_cleaned.groupby('Weather.Condition').size().sort_values(ascending=False)
print("Incidents by Weather Condition:\n", weather_incidents)
```

```python
# Analyze incidents by purpose of flight
purpose_of_flight_incidents = df_aviation_data_2_cleaned.groupby('Purpose.of.flight').size().sort_values(ascending=False)
print("Incidents by Purpose of Flight:\n", purpose_of_flight_incidents)
```

```python
# Calculate the fatality rate per aircraft make/model
df_aviation_data_2_cleaned['Fatality_Rate'] = df_aviation_data_2_cleaned['Total.Fatal.Injuries'] / (df_aviation_data_2_cleaned['Total.Fatal.Injuries'] + df_aviation_data_2_cleaned['Total.Uninjured'])

# Group by Make and Model to find average fatality rate
fatality_rate_by_aircraft = df_aviation_data_2_cleaned.groupby(['Make', 'Model'])['Fatality_Rate'].mean().sort_values().head(10)

print("Aircraft with Lowest Fatality Rate:\n", fatality_rate_by_aircraft)
```

```python
# Example: Assuming 'Aircraft.Age' exists
# Analyze accidents by aircraft age
if 'Aircraft.Age' in df_aviation_data_2_cleaned.columns:
    age_incidents = df_aviation_data_2_cleaned.groupby('Aircraft.Age').size().sort_values(ascending=False).head(10)
    print("Accidents by Aircraft Age:\n", age_incidents)
```

```python
# Analyze risk by number of engines
engine_risk = df_aviation_data_2_cleaned.groupby('Number.of.Engines').size().sort_values(ascending=False)
print("Incidents by Number of Engines:\n", engine_risk)
```

```python
# Analyze destroyed aircraft by make/model
destroyed_aircraft = df_aviation_data_2_cleaned[df_aviation_data_2_cleaned['Aircraft.damage'] == 'Destroyed']
destroyed_by_aircraft = destroyed_aircraft.groupby(['Make', 'Model']).size().sort_values(ascending=False).head(10)

print("Most Destroyed Aircraft by Make/Model:\n", destroyed_by_aircraft)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram of a numerical feature
plt.figure(figsize=(10, 6))
sns.histplot(df_aviation_data_2_cleaned['Total.Fatal.Injuries'])
plt.title('Distribution of Total Fatal Injuries')
plt.xlabel('Total Fatal Injuries')
plt.ylabel('Frequency')
plt.show()

# Boxplot for another feature
plt.figure(figsize=(10, 6))
sns.boxplot(x='Aircraft.Category', y='Total.Fatal.Injuries', data=df_aviation_data_2_cleaned)
plt.title('Fatal Injuries by Aircraft Category')
plt.xlabel('Aircraft Category')
plt.ylabel('Total Fatal Injuries')
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregating the data by 'Purpose.of.flight' and summing up the injury counts
injuries_by_purpose = df_aviation_data_2_cleaned.groupby('Purpose.of.flight')[['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries', 'Total.Uninjured']].sum()

# Plotting a stacked bar chart
injuries_by_purpose.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='viridis')

# Adding titles and labels
plt.title('Injuries Registered by Purpose of Flight', fontsize=16)
plt.xlabel('Purpose of Flight', fontsize=12)
plt.ylabel('Number of Injuries', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Injury Type')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Filter data to include only 'Personal' and 'Business' flights
filtered_df = df_aviation_data_2_cleaned[df_aviation_data_2_cleaned['Purpose.of.flight'].isin(['Personal', 'Business'])]

# Aggregating the data by 'Purpose.of.flight' and summing up the injury counts
injuries_by_purpose = filtered_df.groupby('Purpose.of.flight')[['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries', 'Total.Uninjured']].sum()

# Plotting a stacked bar chart
injuries_by_purpose.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')

# Adding titles and labels
plt.title('Injuries Registered for Personal and Business Flights', fontsize=16)
plt.xlabel('Purpose of Flight', fontsize=12)
plt.ylabel('Number of Injuries', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Injury Type')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Aggregating data by 'Purpose.of.flight' and 'Aircraft.damage'
damage_by_purpose = filtered_df.groupby(['Purpose.of.flight', 'Aircraft.damage']).size().unstack().fillna(0)

# Plotting a stacked bar chart for aircraft damage by flight purpose
damage_by_purpose.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Set2')

# Adding titles and labels
plt.title('Aircraft Damage for Personal and Business Flights', fontsize=16)
plt.xlabel('Purpose of Flight', fontsize=12)
plt.ylabel('Number of Aircrafts Damaged', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Damage Type')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Aggregating data by 'Purpose.of.flight' and 'Injury.Severity'
severity_by_purpose = filtered_df.groupby(['Purpose.of.flight', 'Injury.Severity']).size().unstack().fillna(0)

# Plotting a stacked bar chart for injury severity by flight purpose
severity_by_purpose.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Accent')

# Adding titles and labels
plt.title('Injury Severity for Personal and Business Flights', fontsize=16)
plt.xlabel('Purpose of Flight', fontsize=12)
plt.ylabel('Number of Incidents', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Injury Severity')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Summarizzing my data.
risk_summary = df_aviation_data_2_cleaned[df_aviation_data_2_cleaned['Purpose.of.flight'].isin(['Business', 'Personal'])].groupby('Purpose.of.flight').agg({
    'Total.Fatal.Injuries': ['sum', 'mean', 'median'],
    'Total.Serious.Injuries': ['sum', 'mean', 'median'],
    'Total.Minor.Injuries': ['sum', 'mean', 'median'],
    'Total.Uninjured': ['sum', 'mean', 'median'],
    'Aircraft.damage': 'count'
}).reset_index()

# Display the summary
print(risk_summary)
```

### Business Recommendation

Business flights appear to be less risky in terms of injury severity compared to personal flights. The company should consider prioritizing aircraft for business purposes when making purchasing decisions, as they demonstrate lower injury risk based on historical data.
